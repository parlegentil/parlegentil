## exo 2 : invertion d'une liste

```python
def invertion(tab, n=0):
    """
    
    >>> t = list(range(5))
    >>> invertion(t)
    >>> t
    [4, 3, 2, 1, 0]
    """
    if n == len(tab)//2: return
    #
    tab[n], tab[-1-n] = tab[-1-n], tab[n]
    invertion(tab, n+1)
    
if __name__ == "__main__":
    import doctest
    doctest.testmod()
```
